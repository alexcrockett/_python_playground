```python
import yfinance as yfin
import pandas as pd
```

```python
dat = yfin.Ticker("NOBL")
```

```python
# get historical market data
monthly = dat.history(period='1mo')
```

```python
print(monthly)
```

```python
volume = monthly['Volume']
volume
```

```python
dataset = pd.DataFrame(monthly)
```

```python
high = dataset['High']
low = dataset['Low']
close = dataset['Close']

new_set = pd.concat([high, low, close], axis=1)
print(new_set)
```

```python
import numpy as np
np_set = new_set.to_numpy()
print(np_set)
```
