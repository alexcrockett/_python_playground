```python
import requests
import pandas as pd

# Define the watchlist to work from
watchlist = ['NOBL', 'VYMI', 'PEY', 'XLE', 'VYM', 'HDV', 'SCHD']

# Function to get the API URL for a specific ticker
def stock_api_source(ticker):
	return f'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY&symbol={ticker}&apikey=<Enter your Key Here>'

# Function to get data from API
def data_source(stock_api):
	market_response = requests.get(stock_api)
	data = market_response.json()
	weekly_data = data["Weekly Time Series"]
	return high_low_list(weekly_data)

# Function to create a list with high and low information
def high_low_list(weekly_data):
	new_list = []
	for date, values in weekly_data.items():
		high = float(values['2. high'])
		low = float(values['3. low'])
		new_list.append([date, high, low])
	return new_list

# Function to limit the list to the last 52 weeks
def working_list(new_list):
	return new_list[:52]

# Function to create a DataFrame
def new_frame(wip_list):
	return pd.DataFrame(wip_list, columns=['Date', 'High', 'Low'])

# Function to get the recent high and low prices
def recent_prices(ticker_df):
	recent_high = ticker_df['High'].iloc[0]
	recent_low = ticker_df['Low'].iloc[0]
	return recent_high, recent_low

# Function to get sorted prices
def prices_to_compare(ticker_df):
	sorted_high = ticker_df['High'].sort_values(ascending=False)
	sorted_low = ticker_df['Low'].sort_values(ascending=True)
	return sorted_high, sorted_low

# Function to get the highest and lowest prices
def highest_lowest(sorted_high, sorted_low):
	highest = sorted_high.iloc[0]
	lowest = sorted_low.iloc[0]
	return highest, lowest

# Function to calculate and print high and low prices
def calc_high_low(recent_high, recent_low, highest, lowest):
	if recent_high == highest:
		highest_value = recent_high
		print(f"52 Week High {highest_value}")
	else:
		highest_value = 0
		print(f"52 Week High {highest_value}")

	if recent_low == lowest:
		lowest_value = recent_low
		print(f"52 Week Low {lowest_value}")
	else:
		lowest_value = 0
		print(f"52 Week Low {lowest_value}")

# Main function to run the workflow for each ticker
def main():
	for ticker in watchlist:
		print(f"\nProcessing {ticker}...\n")

		stock_api = stock_api_source(ticker)
		new_list = data_source(stock_api)
		wip_list = working_list(new_list)
		ticker_df = new_frame(wip_list)
		recent_high, recent_low = recent_prices(ticker_df)
		sorted_high, sorted_low = prices_to_compare(ticker_df)
		highest, lowest = highest_lowest(sorted_high, sorted_low)
		calc_high_low(recent_high, recent_low, highest, lowest)

# Run the main function
if __name__ == "__main__":
	main()
```
